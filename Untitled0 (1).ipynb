```python
def extract_keywords(text):
    # Send the text to the OpenAI API to generate keywords
    response = openai.Completion.create(
        engine="davinci",
        prompt=f"Extract the keywords from the following text:\n{text}\n---\nKeywords:",
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.5,
    )
    
    # Extract the keywords from the API response
    keywords = re.findall(r'"(.+)"', response.choices[0].text)
    
    # Return the list of keywords
    return keywords[0]

background_image=extract_keywords(story);
```
